```python
# Function for comparing different approaches using RandomForestRegressor
def score_dataset(X_train, X_valid, y_train, y_valid, estimators):
    model = RandomForestRegressor(n_estimators=estimators, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

# Mean Imputation

```python
def find_columns_with_missing_values(df):
    return [col for col in df.columns if df[col].isnull().any()]
    

# without function finding columns with missing values
cols_with_missing_X_train_numeric = [col for col in X_train_numeric.columns if X_train_numeric[col].isnull().any()]
cols_with_missing_X_valid_numeric = [col for col in X_valid_numeric.columns if X_valid_numeric[col].isnull().any()]

print(cols_with_missing_X_train_numeric, cols_with_missing_X_valid_numeric)

# with function finding columns with missing values
print(find_columns_with_missing_values(X_train_numeric))
print(find_columns_with_missing_values(X_valid_numeric))
```

```python
from sklearn.impute import SimpleImputer

# Imputation
my_imputer = SimpleImputer()
imputed_X_train_numeric = pd.DataFrame(my_imputer.fit_transform(X_train_numeric))
imputed_X_valid_numeric = pd.DataFrame(my_imputer.transform(X_valid_numeric))

imputed_X_train_numeric.columns = X_train_numeric.columns
imputed_X_valid_numeric.columns = X_valid_numeric.columns


print(find_columns_with_missing_values(imputed_X_train_numeric))
print(find_columns_with_missing_values(imputed_X_valid_numeric))
# empty lists => imputed with means
```

# Try with XGBoost here !!

```python
from xgboost import XGBRegressor
base_model = XGBRegressor(random_state = 0)
base_model.fit(imputed_X_train_numeric, y_train)

predictions = base_model.predict(imputed_X_valid_numeric)
print(mean_absolute_error(predictions, y_valid))
```

```python
def xgboost_alg(n_estimators, learning_rate, X_train, y_train, X_valid, y_valid):
    my_model = XGBRegressor(n_estimators=n_estimators, learning_rate=learning_rate, random_state=0)
    my_model.fit(X_train, y_train, 
                 early_stopping_rounds=5, 
                 eval_set=[(X_valid, y_valid)], 
                 verbose=False)

    predictions = my_model.predict(X_valid)
    print("Estimators:",n_estimators,", Learning Rate:", learning_rate)
    print("MAE:",mean_absolute_error(predictions, y_valid))
```

```python
LRs = [i * 0.01 for i in range(1, 10, 1)]
for esti in range(200, 3001, 100):
    for lr in LRs:
        xgboost_alg(esti, lr, imputed_X_train_numeric, y_train, imputed_X_valid_numeric, y_valid)
```

# Using Original Encoding for categorical values

```python
print(find_columns_with_missing_values(X_valid_categorical))
print(find_columns_with_missing_values(X_train_categorical))
```

```python
print("Train")
set1 = set(find_columns_with_missing_values(X_train_categorical))
print(set1)

print("Valid")
set2 = set(find_columns_with_missing_values(X_valid_categorical))
print(set2)

missing_cols_from_both_train_valid = set1.union(set2)
print(missing_cols_from_both_train_valid)
```

```python
X_train_categorical_missing_value_columns_dropped = X_train_categorical.drop(missing_cols_from_both_train_valid, axis = 1)
X_valid_categorical_missing_value_columns_dropped = X_valid_categorical.drop(missing_cols_from_both_train_valid, axis = 1)

print(find_columns_with_missing_values(X_train_categorical_missing_value_columns_dropped))
print(find_columns_with_missing_values(X_valid_categorical_missing_value_columns_dropped))
# no column with missing categorical values
```

```python
print()
```

```python
from sklearn.preprocessing import OrdinalEncoder

cols = X_train_categorical_missing_value_columns_dropped.columns
# Apply ordinal encoder to each column with categorical data
ordinal_encoder = OrdinalEncoder(handle_unknown='ignore')


encoded_X_train_categorical = X_train_categorical_missing_value_columns_dropped.copy()
encoded_X_valid_categorical = X_valid_categorical_missing_value_columns_dropped.copy()

encoded_X_train_categorical[cols] = ordinal_encoder.fit_transform(X_train_categorical_missing_value_columns_dropped[cols])
encoded_X_valid_categorical[cols] = ordinal_encoder.transform(X_valid_categorical_missing_value_columns_dropped[cols])
```

# Trying with RandomForestRegressor

Using only categorical columns dataframe

that too removed missing value columns

Also, performed ordinal encoding

```python
for estimators in range(500, 2001, 500):
    print(score_dataset(encoded_X_train_categorical, encoded_X_valid_categorical, y_train, y_valid, estimators))
    
# high error rate
```

# With xgboost using only categorical data

```python
LRs = [i * 0.01 for i in range(1, 10, 1)]
for esti in range(200, 1500, 100):
    for lr in LRs:
        xgboost_alg(esti, lr, encoded_X_train_categorical, y_train, encoded_X_valid_categorical, y_valid)
```
